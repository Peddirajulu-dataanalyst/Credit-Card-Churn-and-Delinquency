```python
import pandas as pd 

df_churn = pd.read_csv(r"D:\python project files\credit card churn and deliquency\data\bank_churn\cleaned\BankChurners.csv")
df_uci = pd.read_csv(r"D:\python project files\credit card churn and deliquency\data\default_credit_card_clients\cleaned\default of credit card clients.csv")
```

```python
"""
Feature Engineering + Risk Segmantation
Create Inactivity Risk Flag
"""
df_churn["high_inactive_flag"] = df_churn["Months_Inactive_12_mon"] >= 3
```

```python
#create spend drop Flag
df_churn["spend_decline_flag"] = df_churn["Total_Amt_Chng_Q4_Q1"] < 0.8
```

```python
#create util band
df_churn["util_band"] = pd.cut(df_churn["utilization_ratio"], bins = [0,0.2,0.5,0.8,1],labels = ["Very Low","Low","Medium","High"])
```

```python
#Create Risk Score
df_churn["risk_score"] = (df_churn["high_inactive_flag"].astype(int) + df_churn["spend_decline_flag"].astype(int))
```

```python
# Default credit card clients
# Create high utilization flag
df_uci["high_util_flag"] = df_uci["utilization_ratio"] > 0.8
```

```python
#create payment stress flag
df_uci["payment_gap_flag"] = df_uci["payment_gap"] > 0
```

```python
#Build simple default risk score
df_uci["default_risk_score"] = ((df_uci["max_dpd"] >= 2).astype(int) + df_uci ["high_util_flag"].astype(int) + df_uci["payment_gap_flag"].astype(int))
```

```python
df_uci.rename(columns = {"default payment next month": "default_payment_next_month"},inplace = True)
```
