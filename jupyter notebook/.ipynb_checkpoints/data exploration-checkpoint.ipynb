```python
## loading churn data

import pandas as pd
df_churn = pd.read_csv(r"D:\python project files\credit card churn and deliquency\data\bank_churn\raw\BankChurners.csv")
df_churn.head()
```

```python
## Model A -CHURN + BEHAVIOR(BankChurners)
## clean and standardize

df_churn = df_churn.rename(columns = {
    "CLIENTNUM": "Customer_id",
    "Attrition_Flag": "churn_flag",
    "Credit_Limit": "credit_limit",
    "Avg_Utilization_Ratio": "utilization_ratio",
    "Total_Trans_Amt": "total_spend",
    "Total_Trans_Ct": "transaction_count"
})
```

```python
##convert churn

df_churn["churn_flag"] = df_churn["churn_flag"].map({
    "Existing Customer": 0,
    "Attrited Customer": 1
})
```

```python
##Key Behavioral Metrics

df_churn["high_utilization"] = df_churn["utilization_ratio"] > 0.8
df_churn["spend_drop_flag"] = df_churn["Total_Amt_Chng_Q4_Q1"] < 0.8
df_churn["inactive_flag"] = df_churn["Months_Inactive_12_mon"] > 3
```

```python
## loading data of default of credit card clients

import pandas as pd
df_uci = pd.read_csv(r"D:\python project files\credit card churn and deliquency\data\default_credit_card_clients\raw\default of credit card clients.csv")
df_uci.head()
```

```python
## model b - Delinquency + default(UCI)
## creating Maximum DPD

df_uci["max_dpd"] = df_uci[["PAY_0","PAY_2","PAY_3","PAY_4","PAY_5","PAY_6"]].max(axis = 1)
```

```python
## creating deliquency buckets

df_uci["delinquency_bucket"] = pd.cut(
    df_uci["max_dpd"],
    bins=[-2,0,1,2,10],
    labels=["Current","1M Late","2M Late","3+M Late"]
)
```

```python
## default flag

df_uci["default_flag"] = df_uci["default payment next month"]
```

```python
## utilization proxy

df_uci["utilization_ratio"] = df_uci["BILL_AMT1"]/df_uci["LIMIT_BAL"]
```

```python
#adding time simulation (portfolio monitering)

import numpy as np
df_churn["report_month"] = np.random.choice(
    ["Jan","Feb","Mar","Apr","May","Jun"],
    size = len(df_churn)
)
```
