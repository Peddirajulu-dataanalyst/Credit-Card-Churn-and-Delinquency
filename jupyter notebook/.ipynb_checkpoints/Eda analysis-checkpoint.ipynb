```python

## CHURN BEHAVIOR EDA(BankChurners)
## Basic Structure Check

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_churn = pd.read_csv(r"D:\python project files\credit card churn and deliquency\data\bank_churn\cleaned\BankChurners.csv")
df_uci = pd.read_csv(r"D:\python project files\credit card churn and deliquency\data\default_credit_card_clients\cleaned\default of credit card clients.csv")
```

```python
df_churn.shape
```

```python
df_churn.info()
```

```python
df_churn.describe()
```

```python
##convert churn to binary

df_churn["churn_flag"] = df_churn["Attrition_Flag"].map({
    "Existing Customer": 0,
    "Attrited Customer": 1
})
```

```python
##check churn rate

df_churn["churn_flag"].mean()
```

```python
## spend vs churn

sns.boxplot(x = "churn_flag", y = "total_spend", data = df_churn)
plt.show()

##look for lower spend in churn group
```

```python
##Utilization vs Churn

sns.boxplot(x="churn_flag", y="utilization_ratio",data = df_churn)
plt.show()

## check are churn customers high utilization
```

```python
## Inactivity vs churn

sns.barplot(x ="Months_Inactive_12_mon", y ="churn_flag",data = df_churn)
plt.show

## Look for higher churn in 4+ inactive months
```

```python
df_churn[df_churn["Months_Inactive_12_mon"] == 0] ["churn_flag"].value_counts()
```

```python
df_churn["Months_Inactive_12_mon"].value_counts()
```

```python
##Spend change signal

sns.boxplot(x="churn_flag",y="Total_Amt_Chng_Q4_Q1",data = df_churn)
plt.show()
```

```python
## defaut + delinquency EDA (UCI)

df_uci.shape
df_uci.info
df_uci.describe()
```

```python
##default rate

df_uci["default payment next month"].mean()
```

```python
##DPD vs Default

sns.barplot(x = "max_dpd", y = "default payment next month", data = df_uci)
plt.show()
## you should see Higher DPD > Higher default rate
```

```python
## utilization proxy

sns.boxplot(x = "default payment next month" , y = "utilization_ratio", data = df_uci)
plt.show()

##Higher utilization > Higher default
```

```python
## payment gap

df_uci["payment_gap"] = df_uci["BILL_AMT1"] - df_uci["PAY_AMT1"]
sns.boxplot(x = "default payment next month", y ="payment_gap", data = df_uci)
plt.show()
## Large gap  > defaut risk signal
```
